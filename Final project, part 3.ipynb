```python
import pandas as pd
import numpy as np
import ipywidgets
import bqplot
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
old = pd.read_csv("https://data.illinois.gov/datastore/dump/8d9eabcb-08c7-40ae-8252-877fb0e5788d",na_values = {'14b_numHired_AA': 0,
                 '3_headcount':0,
                '14_numHired':0})
old
```

### Data Selection

I first created a pivot table for data visualization to find the African American's hiring facts. For this table, the index is Agency. These Agencies are state entities completing the survey in Illinois. And the column is the Year between 2015 to 2019. Lastly, For the value, I choose the 14_numHired column. According to the websites, this column describes how many employees were hired during the last fiscal year according to the websites? Include new "off the street" hires from the Open Competitive list of eligible and inter and intra agency transfers of current state employees, promotions, voluntary reductions, lateral moves, etc. This can be really helpful.

```python
df=old.pivot(index="Agency",values="14_numHired",columns='Year')
df
```

### First Interactive Visualization

I used the ipywidgets dropdown method to show the pivot table for this interactive visualization. Moreover, I created a sector for the user to click on. This sector included all the agency’s names in it. Users can click one of the agencies, then these codes can provide the number of hiring employees every year.

```python
filtered_df = None

dropdown = widgets.SelectMultiple(
                        options=df.index,
                        description='Sector',
                        disabled=False,
                        layout={'height':'200px', 'width':'60%'})

def filter_dataframe(widget):
    global filtered_df
    selection = list(widget['new'])

    with out:
        clear_output()
        display(df.loc[selection])
        filtered_df = df.loc[selection]

out = widgets.Output()
dropdown.observe(filter_dataframe, names='value')
display(dropdown)
display(out)
```

For the second interactive visualization, it is really similar to the first one. Frist is to create another pivot table. The index and the columns are the same as the first pivot table. ### Third Interactive Visualization

```python
@ipywidgets.interact(Agency_Name=old["Agency"].unique(), style=plt.style.available, nbins = np.arange(20,100))
def make_plot(Agency_Name, style):
    with plt.style.context(style): 
        data = old[old['Agency']== Agency_Name]
        table = data.get(["14b_numHired_AA"])
        #table2= data.get(["Year"])

        
        #fig, ax = plt.subplots(figsize=(12,12))
        plt.figure(figsize=(100,100))
        ax = table.plot(kind="bar")
        plt.title('Afircan American who were hired')
        plt.xlabel('_id')
        plt.ylabel('Number of people hired in different year')

        print(table)
        #plt.show()
```

### Fourth Interactive Visualization

```python
@ipywidgets.interact(Agency_Name=old["Agency"].unique(), style=plt.style.available, nbins = np.arange(20,100))
def make_plot(Agency_Name, style,nbins):
    with plt.style.context(style,nbins): 
        data = old[old['Agency']== Agency_Name]
        table = data.get(["14b_numHired_AA"])
        #table2= data.get(["Year"])

        
        fig, ax = plt.subplots(1,1,figsize=(12,12))
        ax.hist(table,linewidth=0.5, edgecolor="white",bins=nbins) 
        plt.title('Afircan American who were hired')
        plt.xlabel('Number of people hired in different year')

        print(table)
        plt.show()

```

I used a bar chart and a histogram for the third and fourth interactions to show how many African Americans were hired in each different agency.

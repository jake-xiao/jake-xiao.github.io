```python
import pandas as pd
import numpy as np
import ipywidgets
import bqplot 
```

```python
df = pd.read_csv("https://uiuc-ischool-dataviz.github.io/is445_AOUAOG_fall2021/week02/data/building_inventory.csv",na_values = {'Square Footage': 0,
                       'Year Acquired': 0,
                       'Year Constructed': 0})
df
```

```python
old = pd.pivot_table(df, index=["Agency Name"],values=["Square Footage"],columns=["County"],aggfunc=np.mean)

print(old)
```

```python
new=old.columns.levels[1].to_list()
```

```python
old.values
```

# Left Component

```python
heat_map.selected
```

```python
mySelectedLabel = ipywidgets.Label() 
```

```python
col_sc = bqplot.ColorScale(scheme = 'Blues') 
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()


heat_map = bqplot.GridHeatMap(color = np.log10(old.values), row = old.index, column = new,
                            scales={'color':col_sc, 'row':y_sc, 'column':x_sc},
                             interactions={'click':'select'},
                             anchor_style={'fill':'blue'})



c_ax = bqplot.ColorAxis(scale=col_sc, orientation='vertical', side='right')
x_ax = bqplot.Axis(scale=x_sc)
y_ax = bqplot.Axis(scale=y_sc, orientation='vertical')



def on_selected(change): # take a look at what changes when I select a new bin
    if len(change['owner'].selected) == 1: # checking that user is only selecting one element
        i,j = change['owner'].selected[0]
        val = np.mean(old.values)
        mySelectedLabel.value = 'Total mean='+str(val)
        
heat_map.observe(on_selected,'selected')


fig = bqplot.Figure(marks=[heat_map], axes=[c_ax, x_ax, y_ax])

myDashboard = ipywidgets.VBox([mySelectedLabel, fig])
myDashboard
```

# Dashboard


```python
def get_data_value(change):
    i,j = change['owner'].selected[0]
    mask = (df['Agency Name'] == old.index[i]) & \
            (df['County'] == old.columns.levels[1][j])
    # grab buildings subset dataframe
    buildings_subset = df[mask]
    # grab years and sqft from pandas
    grouped = buildings_subset.groupby("Year Acquired")["Square Footage"].sum()
    years = grouped.index
    sqft = grouped.values
    bar.x = years
    bar.y = sqft
heat_map.observe(get_data_value, 'selected')
```

```python
fig.layout.min_width='500px'
fig_bar.layout.min_width = '500px'
figures = ipywidgets.HBox([fig, fig_bar])

myDashboard = ipywidgets.VBox([mySelectedLabel, figures])
myDashboard
```
